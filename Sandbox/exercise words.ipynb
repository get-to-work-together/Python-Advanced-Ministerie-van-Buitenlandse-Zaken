```python
def count_words(tekst):
    tekst = tekst.lower().translate(str.maketrans('', '', string.punctuation))
    woorden = tekst.split()
    unieke_woorden = set(woorden)
    d = {}
    for woord in woorden:
        d[woord] = d.get(woord, 0) + 1
    return d
```
